# 📊 Análise de Vendas - Loja de Informática

**Objetivo:** Analisar 10 vendas para calcular faturamento, ticket médio e identificar produtos mais vendidos.

**Ferramentas:** Python + Pandas

#Importar Bibliotecas

 * Pandas é a ferramenta principal para análise de dados em Python
* Ela nos permite trabalhar com tabelas (DataFrames) de forma eficiente

```python
# Importando a biblioteca pandas
import pandas as pd
```

#Carregar Os Dados

```python
# Lendo o arquivo CSV (planilha) e carregando em um DataFrame
# O CSV contém as colunas: data, produto, quantidade, preco_unitario
df = pd.read_csv('vendas_loja.csv')

# Verificando se os dados foram carregados corretamente
# head() mostra as primeiras 5 linhas do DataFrame
try:
    df = pd.read_csv('vendas_loja.csv')
    print("✅ Arquivo lido com sucesso!")
    print(f"📊 {len(df)} linhas encontradas")
    print(df.head())
except FileNotFoundError:
    print("❌ Arquivo não encontrado! Verifique o nome e a pasta")
```

#Criar Nova Coluna (Faturamento)

```python
# Criando uma nova coluna chamada 'faturamento'
# O faturamento de cada venda é calculado como: quantidade × preço unitário
# Isso é chamado de "feature engineering" - criar novas variáveis a partir das existentes
df['faturamento'] = df['quantidade'] * df['preco_unitario']

# Exibindo os dados com a nova coluna
# As colunas agora são: data, produto, quantidade, preco_unitario, faturamento
df.head()
```

# Métricas de Negócio

```python
# ===== ANÁLISES ESTATÍSTICAS BÁSICAS =====
# Essas métricas são essenciais para qualquer relatório de vendas

# 1. Total de vendas (número de transações)
total_vendas = len(df)
print(f" Total de transações: {total_vendas}")

# 2. Faturamento total (soma de todas as vendas)
# O formatador :,.2f adiciona separador de milhar e 2 casas decimais
faturamento_total = df['faturamento'].sum()
print(f" Faturamento total: R$ {faturamento_total:,.2f}")

# 3. Ticket médio (média do faturamento por venda)
# Indica o valor médio gasto por cliente em cada compra
ticket_medio = df['faturamento'].mean()
print(f" Ticket médio por venda: R$ {ticket_medio:,.2f}")

# 4. Produto mais vendido (em quantidade em UMA única venda)
# idxmax() encontra o índice (posição) onde está o maior valor
indice_maior = df['quantidade'].idxmax()
produto_destaque = df.loc[indice_maior, 'produto']
quantidade_destaque = df['quantidade'].max()
print(f" Produto com maior quantidade em uma venda: {produto_destaque} ({quantidade_destaque} unidades)")
```

# Faturamento por Produto

```python
# ===== ANÁLISE DE FATURAMENTO POR CATEGORIA =====
# O groupby agrupa os dados por produto e soma o faturamento de cada grupo
# Isso responde: "Qual produto gerou mais receita total?"

faturamento_por_produto = df.groupby('produto')['faturamento'].sum()

# Ordenando do maior para o menor (para facilitar a leitura)
faturamento_ordenado = faturamento_por_produto.sort_values(ascending=False)

print(" FATURAMENTO TOTAL POR PRODUTO:")
print(faturamento_ordenado)

# Identificando o produto campeão de faturamento
produto_campeao = faturamento_ordenado.index[0]
valor_campeao = faturamento_ordenado.iloc[0]
print(f"\n Produto que mais faturou: {produto_campeao} (R$ {valor_campeao:,.2f})")
```

# Filtros

```python
# ===== FILTROS =====
# Filtros são usados para responder perguntas específicas sobre os dados
# A lógica é: df[condição] onde condição é uma comparação em uma coluna

print("=" * 50)
print(" PERGUNTA 1: Quais vendas foram do produto Notebook?")
print("=" * 50)
vendas_notebooks = df[df['produto'] == 'Notebook']
print(vendas_notebooks)

print("=" * 50)
print(" PERGUNTA 2: Quais vendas tiveram quantidade maior que 2?")
print("=" * 50)
vendas_quantidade = df[df['quantidade'] > 2]
print(vendas_quantidade)

print("=" * 50)
print(" PERGUNTA 3: Quais vendas superaram R$ 1.000?")
print("=" * 50)
vendas_1000 = df[df['faturamento'] > 1000]
print(vendas_1000)
```

# Salvando Resultados

```python
# ===== EXPORTANDO RESULTADOS =====
# Salvando as análises em arquivos CSV para compartilhar
# index=False evita criar uma coluna extra de índices

# 1. Resumo do faturamento por produto
faturamento_por_produto.to_csv('resumo_faturamento.csv', header=['faturamento'])

# 2. Vendas acima de R$ 1000 (clientes premium)
vendas_1000.to_csv('vendas_premium.csv', index=False)

print(" Arquivos salvos com sucesso!")
print(" resumo_faturamento.csv - Faturamento por produto")
print(" vendas_premium.csv - Vendas acima de R$ 1.000")
```
